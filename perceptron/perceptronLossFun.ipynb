```python
def perceptron(X, y):
    w1 = w2 = b = 1.0
    lr = 0.1
    epochs = 1000

    for _ in range(epochs):
        errors = 0
        for i in range(X.shape[0]):
            z = w1*X[i][0] + w2*X[i][1] + b
            if z * y[i] <= 0:
                w1 += lr * y[i] * X[i][0]
                w2 += lr * y[i] * X[i][1]
                b  += lr * y[i]
                errors += 1
        if errors == 0:
            break

    return w1, w2, b
```

```python
w1,w2,b = perceptron(X,y)
```
