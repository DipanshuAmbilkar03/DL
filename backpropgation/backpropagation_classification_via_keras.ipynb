```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,12,0]],columns=['cgpa','profile_score','placed'])
```

```python
df
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(2,activation='sigmoid',input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
new_weights = [
    np.array([[0.1,0.1],[0.1,0.1]], dtype='float32'),
    np.array([0.,0.] , dtype='float32'),
    np.array([[0.1],[0.1]], dtype='float32'),
    np.array([0.1] , dtype='float32'),
]
```

```python
new_weights
```

```python
model.set_weights(new_weights)
```

```python
model.get_weights()
```

```python
optimize = keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimize,loss='binary_crossentropy')
```

```python
model.fit(df.iloc[:,0:-1].values,df['placed'].values,epochs=75,verbose=1,batch_size=1)
```
